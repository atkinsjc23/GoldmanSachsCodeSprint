Mike is a stock trader and makes a profit by buying and selling stocks. He buys a stock at a lower price and sells it at a higher price to book a profit. He has come to know the stock prices of a particular stock for n upcoming days in future and wants to calculate the maximum profit by doing the right transactions (single transaction = buying + selling). Can you help him maximize his profit?

A transaction starts after the previous transaction has ended. Two transactions can't overlap or run in parallel. The stock prices are given in the form of an array A for n days. Given the stock prices and a positive integer k, find and print the maximum profit Mike can make in at most k transactions. For each query print the maximum profit earned by Mike on a new line.

```python
import sys


def oneTrans(l,m):
    ans=0
    for i in range (0,m-1):
        for j in range (i+1,m):
            if (l[j]-l[i])>ans:
                ans=(l[j]-l[i])
    return(ans)

def twoTrans(l,m):
    ans=0
    for i in range (0,m-3):
        for j in range (i+1,m-2):
            for k in range (j+1,m-1):
                for o in range (k+1,m):
                    if (l[j]-l[i]+l[o]-l[k])>ans:
                        ans=(l[j]-l[i]+l[o]-l[k])
    return(ans)

def threeTrans(l,m):
    ans=0
    for i in range (0,m-5):
        for j in range (i+1,m-4):
            for k in range (j+1,m-3):
                for o in range (k+1,m-2):
                    for p in range (o+1,m-1):
                        for q in range (p+1,m):
                            if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p])>ans:
                                ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p])
    return(ans)

def fourTrans(l,m):
    ans=0
    for i in range (0,m-7):
        for j in range (i+1,m-6):
            for k in range (j+1,m-5):
                for o in range (k+1,m-4):
                    for p in range (o+1,m-3):
                        for q in range (p+1,m-2):
                            for r in range (q+1,m-1):
                                for s in range (r+1,m):
                                    if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r])>ans:
                                        ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r])
    return(ans)

def fiveTrans(l,m):
    ans=0
    for i in range (0,m-9):
        for j in range (i+1,m-8):
            for k in range (j+1,m-7):
                for o in range (k+1,m-6):
                    for p in range (o+1,m-5):
                        for q in range (p+1,m-4):
                            for r in range (q+1,m-3):
                                for s in range (r+1,m-2):
                                    for t in range (s+1,m-1):
                                        for u in range (t+1,m):
                                            if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t])>ans:
                                                ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t])
    return(ans)

def sixTrans(l,m):
    ans=0
    for i in range (0,m-11):
        for j in range (i+1,m-10):
            for k in range (j+1,m-9):
                for o in range (k+1,m-8):
                    for p in range (o+1,m-7):
                        for q in range (p+1,m-6):
                            for r in range (q+1,m-5):
                                for s in range (r+1,m-4):
                                    for t in range (s+1,m-3):
                                        for u in range (t+1,m-2):
                                            for v in range (u+1,m-1):
                                                for w in range (v+1,m):
                                                    if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v])>ans:
                                                        ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v])
    return(ans)

def sevenTrans(l,m):
    ans=0
    for i in range (0,m-13):
        for j in range (i+1,m-12):
            for k in range (j+1,m-11):
                for o in range (k+1,m-10):
                    for p in range (o+1,m-9):
                        for q in range (p+1,m-8):
                            for r in range (q+1,m-7):
                                for s in range (r+1,m-6):
                                    for t in range (s+1,m-5):
                                        for u in range (t+1,m-4):
                                            for v in range (u+1,m-3):
                                                for w in range (v+1,m-2):
                                                    for x in range (w+1,m-1):
                                                        for y in range (x+1,m):
                                                            if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v]+l[y]-l[x])>ans:
                                                                ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v]+l[y]-l[x])
    return(ans)

def eightTrans(l,m):
    ans=0
    for i in range (0,m-15):
        for j in range (i+1,m-14):
            for k in range (j+1,m-13):
                for o in range (k+1,m-12):
                    for p in range (o+1,m-11):
                        for q in range (p+1,m-10):
                            for r in range (q+1,m-9):
                                for s in range (r+1,m-8):
                                    for t in range (s+1,m-7):
                                        for u in range (t+1,m-6):
                                            for v in range (u+1,m-5):
                                                for w in range (v+1,m-4):
                                                    for x in range (w+1,m-3):
                                                        for y in range (x+1,m-2):
                                                            for z in range (y+1,m-1):
                                                                for a in range (z+1,m):
                                                                    if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v]+l[y]-l[x]+l[a]-l[z])>ans:
                                                                        ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v]+l[y]-l[x]+l[a]-l[z])
    return(ans)

def nineTrans(l,m):
    ans=0
    for i in range (0,m-17):
        for j in range (i+1,m-16):
            for k in range (j+1,m-15):
                for o in range (k+1,m-14):
                    for p in range (o+1,m-13):
                        for q in range (p+1,m-12):
                            for r in range (q+1,m-11):
                                for s in range (r+1,m-10):
                                    for t in range (s+1,m-9):
                                        for u in range (t+1,m-8):
                                            for v in range (u+1,m-7):
                                                for w in range (v+1,m-6):
                                                    for x in range (w+1,m-5):
                                                        for y in range (x+1,m-4):
                                                            for z in range (y+1,m-3):
                                                                for a in range (z+1,m-2):
                                                                    for b in range (a+1,m-1):
                                                                        for c in range (b+1,m):
                                                                            if (l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v]+l[y]-l[x]+l[a]-l[z]+l[c]-l[b])>ans:
                                                                                ans=(l[j]-l[i]+l[o]-l[k]+l[q]-l[p]+l[s]-l[r]+l[u]-l[t]+l[w]-l[v]+l[y]-l[x]+l[a]-l[z]+l[c]-l[b])
    return(ans)

def peakValley(A):
    #This function outputs a new array of local extrema from the array A and initialized with first entry of A
    incFlag=False
    decFlag=False
    new=[A[0]]
    for i in range(len(A)-1):
        if A[i]<=A[i+1]:
            if not incFlag:
                new.append(A[i+1])
                incFlag=True
                decFlag=False
            else:
                new.pop()
                new.append(A[i+1])
        if A[i]>=A[i+1]:
            if not decFlag:
                new.append(A[i+1])
                decFlag=True
                incFlag=False
            else:
                new.pop()
                new.append(A[i+1])
    return(new)

def traderProfit(k, n, A):
    l=peakValley(A)
    m=len(l)
    
    #determine the number of transactions
    if (m%2)==0:
        if l[0]<l[1]:
            if k>(m//2):
                k=(m//2)
        else:
            if k>(m//2)-1:
                k=(m//2)-1
    else:
        if k>(m//2):
            k=(m//2)
            
    #compute maximum profit
    if k==1:
        return(oneTrans(l,m))
    elif k==2:
        return(twoTrans(l,m))
    elif k==3:
        return(threeTrans(l,m))    
    elif k==4:
        return(fourTrans(l,m))
    elif k==5:
        return(fiveTrans(l,m))
    elif k==6:
        return(sixTrans(l,m))   
    elif k==7:
        return(sevenTrans(l,m))
    elif k==8:
        return(eightTrans(l,m))
    elif k==9:
        return(nineTrans(l,m))
    else:
        return(0)
    
    
if __name__ == "__main__":
    q = int(input().strip())
    for a0 in range(q):
        k = int(input().strip()) %0<k<10, At most nine transactions
        n = int(input().strip())
        arr = list(map(int, input().strip().split(' ')))
        result = traderProfit(k, n, arr)
        print(result)
```
